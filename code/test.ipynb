```python
import numpy as np

nhex = 1338
filename = 'data/trip_time_od_hex.csv'
#process the line based file into a hour by
data=np.zeros((24, nhex,nhex))
with open(filename,'r') as f:
    next(f)
    for lines in f:
        line=lines.strip().split(',')
        h,o,d,t=line[1:] # hour, oridin, dest, number of trip
        data[int(h),int(o),int(d)]=float(t)
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

df = pd.read_csv('logs/parsed_results_new.csv')
df['time'] = (df['time']-df['time'][0])//(60)

df = df.set_index('time')
fig, axes = plt.subplots(nrows=3, ncols=3,figsize=(12,15))
axe = axes.ravel()
# print(
for id,col in enumerate(df.columns.tolist()[1:10]):
    print(id,col)
    df["%s"%col].plot(ax=axe[id])    
    axe[id].set_ylabel(col)
    axe[id].set_title(col)
    axe[id].set_xlabel('time(min)')

plt.show()
```
